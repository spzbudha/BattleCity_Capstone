## Introduction: Business Problem <a name="introduction"></a>

In this project we will try assist a business woman (Restaurant Owner) of South African descend, who at her own capacity narrowed out two cities out of many that she would like to visit. Bearing in mind that her final decision between the two destinations will need to satisfy her long time conquest of encorporating various cuisines to her african food chain of restaurants upon her return.
Our aim is to help her make an informed decision between New York City and Toronto, specifically.
